```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 데이터

```python
df=pd.read_csv('glass.csv')
df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

## 결측치 확인

```python
df = df.dropna() # 결측치가 포함된 행을 삭제
df.isnull().sum(axis=0)
```

## 피처, 레이블 분리

```python
X = df.drop(['Type'], axis=1)
y = df['Type']
X.head()
```

## 트레이닝, 테스트 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, random_state=42) # random_state : seed값

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

## 데이터 스케일링

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
y_train = y_train.values
X_test = sc.transform(X_test)
y_test = y_test.values
```

# 학습

```python
from sklearn.metrics import *

def print_scores(y_true, y_pred):
  print('accuracy_score: {0:.4f}'.format(accuracy_score(y_true, y_pred)))
  print('precision_score: {0:.4f}'.format(precision_score(y_true, y_pred, average='macro')))
  print('recall_score: {0:.4f}'.format(recall_score(y_true, y_pred, average='macro')))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print_scores(y_test, y_pred)
```

## SVM

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print_scores(y_test, y_pred)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
print_scores(y_test, y_pred)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

dtc = RandomForestClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
print_scores(y_test, y_pred)
```
