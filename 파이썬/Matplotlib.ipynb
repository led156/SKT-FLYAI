# 2. 기본 그래프 그리기

### plot() 함수

- 수치 데이터를 설정하면 선 그래프(Line graph)를 그린다.
- 그래프를 화면에 표시하기 위해 plt.show() 함수를 호출한다.
- 그래프의 타이틀, X축, Y축의 이름을 설정할 수 있다

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 10)**2
plt.plot(x)
plt.show()
```

```python
x = np.arange(10, 20, 0.1)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 

x = np.arange(0, 1, 0.1)
y = x**2
plt.plot(x, y)

plt.title("제목")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 여러 개의 그래프 그리기



```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

### 그래프 스타일 지정

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)

plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'b.')
plt.plot(x, y3, 'g:^')

plt.show()
```

```python
x = np.arange(10)
plt.plot(x, x, 'r.--')
plt.plot(x, x+2, 'bo-')
plt.plot(x, x+4, 'gv-.')
plt.show()
```

```python
x = np.arange(10)
plt.plot(x, x, color = "r")
plt.plot(x, x+2, color = "green")
plt.plot(x, x+4, color = "#524FA1")
plt.plot(x, x+6, 'k-')
plt.show()
```

### 그래프 형상 (figure) 설정
https://codetorial.net/matplotlib/set_legend.html

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x+0.5)
y3 = np.sin(x+1)

plt.plot(x, y1, 'r--', label = 'sin A')
plt.plot(x, y2, 'g-.', label = 'sin B')
plt.plot(x, y3, 'b', label = 'sin C')

plt.xlim(0, 20)

plt.legend(loc='lower right') 
# loc, location : upper/center/lower, left/right | (pos1, pos2) | best
# ncol = #, fontsize = #
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, 'b-.', linewidth=3)
plt.xticks(np.arange(0, 10, 2)) # -> 0, 2, 4, 6, 8, 10
plt.yticks(np.arange(-1.0, 1.1)) # -> -1.0 ~ 1.1
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, linewidth=3)
plt.xticks([0, 5, 10], labels=['Jan', 'Feb', 'Mar'], rotation=45) # 0 = Jan, 5 = Feb, 10 = Mar
# rotation : 회전
plt.yticks([-1, 0, 1], labels=['min', 'ave', 'max'])
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, linewidth=3)
plt.xticks([0, 5, 10], labels=['2022-01-01', '2022-01-02', '2022-01-03'], rotation=45)
plt.yticks([-1, 0, 1], labels=['2022-01-01', '2022-01-02', '2022-01-03'])
plt.show()
```

```python
# 눈금 삭제하기
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, linewidth=3)
plt.xticks([])
plt.yticks([])
plt.show()
```

### 그래프 구조


```python
x = np.arange(1, 30)
y = np.log(x)

plt.figure()
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(1, 30)
y = np.log(x)

plt.figure(figsize = (8,3))
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(1, 30)
y = np.log(x)

fig = plt.gcf()
fig.set_size_inches(8, 3)
plt.plot(x, y)
plt.show()
```

### 그래프 여러 개 그리기

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1) 
plt.plot(x, y, 'r-')
plt.subplot(2, 2, 2)
plt.plot(x, z, 'g--')
plt.subplot(2, 2, 3)
plt.plot(x, z, 'b-.')

plt.show()
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(2, 2)
fig.set_size_inches(10, 8)

ax[0][0].plot(x, y, 'r-')
ax[0][1].plot(x, z, 'g-')
ax[1][0].plot(x, z, 'b-')
plt.show()
```
